 question--> 1. What is the difference between a function and a method in Python?

 Function:

question--> 3. What are the different ways to define and call a function in Python?
answer--> In Python, functions can be defined and called in multiple ways, depending on the use case. Below is a detailed guide to the different ways to define and call functions:
1. Standard Function Definition
Defined using def with optional parameters.

Define:
def greet(name):
  print(f"Hello, {name}!")

Call:
greet("Alice")
2. Function with Default Parameters
You can provide default values to parameters.

Define:
def greet(name="Guest"):
    print(f"Hello, {name}!")
Call:
greet()         
greet("Ravi")   
3. Function with *args (Variable Positional Arguments)
Allows passing multiple positional arguments.
Define:
def show_skills(*skills):
    for skill in skills:
        print(skill)
Call:
show_skills("Python", "SQL", "ML")

4. Function with **kwargs (Variable Keyword Arguments)
Allows passing multiple keyword arguments.

Define:
def show_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")
Call:
show_info(name="Alice", age=25, city="Delhi")

5. Function Returning a Value
Can return data using return.
Define:
def add(a, b):
    return a + b
Call:
result = add(3, 5)
print(result)

6. Lambda Function (Anonymous Function)
Used for small, single-expression functions.

Define & Call:
square = lambda x: x ** 2
print(square(4))

7. Recursion (Function Calling Itself)
A function can call itself.
Define:
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)
Call:
print(factorial(5))

8. Function as an Argument
Functions can be passed as arguments.

Define:
def apply(func, x):
    return func(x)
def double(n):
  return n * 2
Call:
print(apply(double, 5))

question--> 6 . Explain the concept of generators in Python and how they are defined.

answer--> A generator is a special type of iterator in Python that is used to generate values on the fly, instead of storing them all in memory at once. This makes generators memory-efficient, especially when dealing with large datasets or infinite sequences.

Concept Behind Generators
Generators use the yield keyword instead of return.
Each time yield is called, the function's state is saved, and execution is paused.
When next() is called again, it resumes from where it left off.

Key Features of Generators-->
Feature	and Description:
Uses yield-->	Replaces return and allows function to resume later
Lazy Evaluation-->	Generates values one at a time, only as needed
Memory Efficient-->	Doesn’t store all values in memory
Iterable-->	Yes,can be used in a for loop or with next()
Stops with Exception-->	Raises StopIteration when exhausted


question-->7  What are the advantages of using generators over regular functions?

answer--> Using generators over regular functions (especially ones that return lists or other collections) offers several key advantages, particularly in terms of performance and memory usage.
Advantages of Generators in Python
1.Memory Efficiency
Generators yield one value at a time, instead of storing the entire result in memory.

2.Faster Start Time
Generators don't compute all values upfront — they produce values on demand, which means they start faster and can be more responsive.

3.Lazy Evaluation
Values are only computed when needed, which is great for:
Processing large files line by line
Streaming data
Working with infinite sequences

4.Pipeline Support
Generators can be chained together to form efficient data processing pipelines.

5.Avoids Temporary Storage
Generators do not require temporary storage for intermediate results, reducing disk or memory I/O.

6.Clean Syntax with yield
Generator functions are easier to write and read compared to manually implementing an iterator class with __iter__() and __next__() methods.

question--> 9  Explain the purpose and usage of the `map()` function in Python.

answer--> The map() function in Python is used to apply a function to every item in an iterable (like a list or tuple) and return a new map object (which is an iterator).

Purpose:
To transform data efficiently by applying a given function to all elements of an iterable — without writing a loop manually.
function: A function (often a lambda) to apply.
iterable: A sequence like a list, tuple, or string.
map() returns a map object, which is an iterator. You can convert it to a list, tuple, etc., to view the result.
If you need to apply simple transformations to elements, map() is cleaner and faster.
For complex logic, use a for loop or list comprehension for better readability.





question--> 10 What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

answer--> The built-in functions map(), filter(), and reduce() in Python are powerful tools from functional programming used to process collections like lists. Here's a breakdown of each and how they differ:

1. map() — Transform Each Item
Purpose: Applies a function to each item in an iterable and returns a new iterable (a map object).

2. filter() — Select Items That Meet a Condition
Purpose: Filters the items in an iterable based on a function that returns True or False.

3. reduce() — Reduce to a Single Value
Purpose: Applies a function cumulatively to the items of an iterable, reducing the iterable to a single value.

```python
#question-->4  Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Example usage:
number = 29
if is_prime(number):
    print(f"{number} is a prime number.")
else:
    print(f"{number} is not a prime number.")
```

```python
#question-->6  Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage:
max_exponent = 5
print("Powers of 2 up to exponent", max_exponent, ":")
for power in powers_of_two(max_exponent):
    print(power)
```

```python
#question--> 8 Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

# Sample list of tuples
data = [(1, 'banana'), (2, 'apple'), (3, 'cherry'), (4, 'date')]

# Sorting the list of tuples based on the second element
sorted_data = sorted(data, key=lambda x: x[1])

# Example usage
print("Sorted list of tuples based on the second element:")
print(sorted_data)
```

```python
#question--> 9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit..

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 37, 100]

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Example usage
print("Celsius temperatures:", celsius_temperatures)
print("Fahrenheit temperatures:", fahrenheit_temperatures)
```

```python
#question--> 10 Create a Python program that uses `filter()` to remove all the vowels from a given string.

def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = "Hello, World!"

filtered_string = ''.join(filter(is_not_vowel, input_string))

# Example usage
print("Original string:", input_string)
print("String without vowels:", filtered_string)
```
